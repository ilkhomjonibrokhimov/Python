## Defining Functions

```python
def greet():
    print("Hello world")

greet()
```

```python
def introduce(name):
    print("Hello", name)

introduce("John")
```

```python
def add(a, b):
    return a + b

result = add(2, 3)
print(result)
```

```python
def square(n):
    """Returns the square of a number"""
    return n * n

square(4)
```

```python
def greeting(name, message):
    print(f"{message}, {name}!")

greeting("John", "Hello")
```

```python
greeting(message="Hello", name="John")
```

```python
def pos_only(x, /, y):
    return x + y

print(pos_only(3, y=4))
```

```python
def key_only(*, x, y):
    return x * y

print(key_only(x=3, y=4))
```

```python
def greet(name="Guest"):
    print(f"Hello, {name}")

greet("Alice")
greet()
```

```python
def power(base, exponent=2):
    return base ** exponent

print(power(2))
print(power(2, 3))
```

```python
def sum_all(*args):
    return sum(args)

print(sum_all(1,2 ,3, 4,4,5))
```

```python
def print_info(**kwarg):
    print(kwarg)

print_info(name = "Alice", age = 20, city = "New York")
```

```python
square = lambda x: x ** 2
print(square(2))
```

```python
names = ["Charlie", "Alice", "Bob", "David"]
names.sort()
names.sort(key=lambda x: len(x))
print(names)
```

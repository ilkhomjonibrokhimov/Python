## Decorators

```python
def print_hello():
    print('hello')

def print_another():
    print("Another function")

a = print_hello
a()
```

```python
def decorator1(func):
    print("BEFORE")
    func()
    print("AFTER")

decorator1(print_hello)
print()
decorator1(print_another)
```

```python
def decorator(func):
    def wrapper():
        print("Befor function call")
        func()
        print("After function call")
    return wrapper
```

```python
var = decorator(print_hello)
var()
```

```python
print_hello = decorator(print_hello)
```

```python
print_hello()
```

```python
@decorator
def say_hello():
    print("Hello")

say_hello.__name__
say_hello()
```

```python
def bold(func):
    def wrapper():
        return f"<b>{func()}.</b>"
    return wrapper

@bold
def greet():
    return "Hello"

print(greet())
```

```python
def repeat(n):
    def decorator(func):
        def wrapper(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return wrapper
    return decorator

@repeat(4)
def greet(name):
    print(f"Hello, {name}!")

greet("Alice")
```

```python
repeat(1)(greet)("John")
```

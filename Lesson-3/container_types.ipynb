```python
list('asjdbjdf')
```

```python
a = [1, 2, 3]
b = a
a.append(4)

print(a)
print(b)
```

```python
a = [1, 2, 3]
b = a.copy()

print(id(a))
print(id(b))
```

```python
x = [[1, 2], [3, 4]]
y = x.copy()

x[0].append(5)

print(x)
print(y)
```

```python
from copy import copy, deepcopy

x = [[1, 2], [3, 4]]
y = deepcopy(x)

x[0].append(5)

print(x)
print(y)
```

## Working with files

```python
file = open('sample.txt', mode = 'r')
file.read()
```

```python
file = open('sample.txt', mode = 'w')
file.write("World")
file.close()
```

```python
file = open('sample.txt', mode = 'a')
file.write(" apple")
file.close()
```

```python
with open("sample.txt") as f:
    print(f.read())

f.read()
```

```python
with open('sample.txt') as f:
    txt = f.readline()
    print(type(txt))
    print(txt)
    txt = f.readline()
    print(type(txt))
    print(txt)
    
```

```python
with open('sample.txt') as f:
    for line in f:
        print(line, end='')
    
```

```python
with open('sample.txt') as f:
    file_contents = f.read(100)
    print(file_contents)
```

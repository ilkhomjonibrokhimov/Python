# List Comprehensions

```python
numbers = [2, 3, 4, 5]
squares = []
for number in numbers:
    squares.append(number ** 2)
print(squares)
```

```python
numbers = [2, 3, 4, 5]
squares = [val**2 for val in numbers]
print(squares)
```

```python
numbers = [2, 3, 4, 5]
squares = []
for number in numbers:
    if number%2 == 0:
        squares.append(number ** 2)
print(squares)
```

```python
numbers = [2, 3, 4, 5]
squares = [val**2 for val in numbers if val % 2 == 0]
print(squares)
```

```python
a =[]
for i in range(3):
    for j in range(3):
        a.append((i, j))
    
print(a)
```

```python
a = [(x, y) for x in range(3) for y in range(3)]
print(a)
```

```python
matrix = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
[i for row in matrix for i in row]
```

Dictionary Comprehensions:

```python
keys = ['a', 'b', 'c', 'd', 'e']
values = [1, 2, 3, 4, 5]
d = {k:v for (k, v) in zip(keys, values)}
print(d)
```

```python
reverse = {v:k for (k, v) in d.items()}
print(reverse)
```

```python
numbers = [1,2, 3, 4, 5]
squares = {num:num**2 for num in numbers}
print(squares)
```

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
matrix_squares = {pos:{num:num**2 for num in rows} for pos, rows in enumerate(matrix)}
print(matrix_squares)
```

Generator expressions:

```python
sum_even = sum(num**2 for num in range(10) if num % 2 == 0)
print(sum_even)
```

```python
import numpy as np
```

```python
a = np.arange(50)
b = np.random.randint(0, 50, 50)
sizes = np.abs(np.random.randn(50))*100
plt.plot(a, b)
```

```python
plt.scatter(a, b, s=sizes, c=b, cmap='cool')
plt.colorbar()
```

```python
data = np.random.randn(1000)
plt.hist(data, bins=20, color='skyblue', edgecolor='k')
plt.title('Basic Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## Loops

```python
n = 10
while n <= 20:
    print("Executed n =", n)
    n += 1
```

```python
while True:
    print("Still running")
    break
```

```python
cnt = 1

while True:
    print("Cnt =", cnt)
    cnt += 1
    if cnt == 10:
        break
```

```python
cnt = 0

while cnt < 10:
    cnt += 1
    if cnt == 5:
        continue # when this condition is the case, loop goes back to the beginning, ignoring further commands.
    print("Cnt =", cnt)
```

```python
l = [1, 2, 3]
i = 0
check_number = 3
while i < len(l):
    if l[i] == check_number:
        print("Found")
        break
    i += 1
else:
    print("Not found")
```

## FOR loop

```python
lst = [1, 2, 3, 4,]
for a in lst:
    print(a)
```

```python
users = {"user1": "John", "user2": "Adam", "user3": "Josh"}
for user in users:
    print(user, users[user])
```

```python
users = {"user1": "John", "user2": "Adam", "user3": "Josh"}
for user in users.values():
    print(user)
```

```python
users = [("user1", "John"), ("user2", "Adam"), ("user3", "Josh")]

for user in users:
    print(user)
```

```python
users = [("user1", "John"), ("user2", "Adam"), ("user3", "Josh")]

for key, user in users:
    print(user)
```

```python
languages = ["Java", "Python", "JS"]
versions = [14, 3, 6]

for i in range(len(languages)):
    print(languages[i], versions[i])
```

```python
for i in zip(languages, versions):
    print(i)
```

```python
for language, version in zip(languages, versions):
    print(language, version)
```

```python
for i in range(3):
    for j in range(3):
        print(i, j)
```

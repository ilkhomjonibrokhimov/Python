Raise exception

```python
x = -1

if x < 0:
    raise Exception("Sorry, no number below than zero")
```

```python
x = 'hello'

if not type(x) is int:
    raise TypeError("Only integers are allowed")
```

```python
try:
    raise ImportError("Bad import")
except ImportError as error:
    print(error)
    print(type(error))
    print(error.args)
```

```python
try:
    import something
except ImportError as e:
    print(e)
```

```python
try:
    a = 1 / 0
except ZeroDivisionError:
    print("Cannot divide")
else:
    print(a)
finally:
    print("finally executed")
```

```python
try:
    a = 1 / 1
except ZeroDivisionError:
    print("Cannot divide")
else:
    print(a)
finally:
    print("finally executed")
```

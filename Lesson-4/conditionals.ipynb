- True and True = True
- True and False = False
- True or False = True

```python
age = 18
car = 'BMW'

if age >= 18 and car in ['Honda', 'BMW', 'Toyota']:
    print(True)
else:
    print(False)
```

```python
age = 40

if age < 0:
    print("Incorrect number for age. Age can't be negative.")
elif age < 18:
    print("You can't vote.")
elif age < 150:
    print("You can vote.")
else:
    print("Incorrect number for age. Humans can't live that long.")
```
